# HW 2 หา pattern ที่น่าสนใจใน Dataset หนังอินเดีย (india imdb)

```python
# prompt: read file IMDb Movies India.csv from my drive

drive.mount('/content/drive/')
path = '/content/drive/My Drive/Dataset'

india_data = pd.read_csv(os.path.join(path, 'IMDb Movies India (1).csv'), encoding='latin-1')
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Cleaning and Standardizing Column Values
india_data.columns = india_data.columns.str.strip().str.lower()
```

```python
# Drop rows with missing 'name' as it is a key identifier
india_data = india_data.dropna(subset=['name'])

# Fill missing numerical columns with mean, using india_data's mean
india_data['rating'] = india_data['rating'].fillna(india_data['rating'].mean())
```

```python
# Fill missing categorical columns with 'Unknown'
categorical_cols = ['genre', 'director', 'actor 1', 'actor 2', 'actor 3']
india_data[categorical_cols] = india_data[categorical_cols].fillna('Unknown')
```

```python
# Convert 'year' and 'duration' to numeric by extracting numeric values
# Check if 'year' is already numeric, if not, convert it
if not pd.api.types.is_numeric_dtype(india_data['year']):
    india_data['year'] = india_data['year'].astype(str).str.extract('(\d{4})').astype(float)
else:
    print("Year column is already numeric.")

# Similarly, check if 'duration' is already numeric, if not, convert it
if not pd.api.types.is_numeric_dtype(india_data['duration']):
    india_data['duration'] = india_data['duration'].astype(str).str.extract('(\d+)').astype(float)
else:
    print("Duration column is already numeric.")
```

```python
# Convert 'votes' to numeric
india_data['votes'] = pd.to_numeric(india_data['votes'], errors='coerce')
```

```python
# Fill missing values in numeric columns with median
for col in ['year', 'duration', 'votes']:
    india_data[col] = india_data[col].fillna(india_data[col].median())
```

```python
# Step 2: Data Integration
# Create a new feature for decade
india_data['decade'] = (india_data['year'] // 10 * 10).astype(int)
```

```python
# Step 3: Dimensionality Reduction using PCA
numerical_cols = ['year', 'duration', 'rating', 'votes']

# Standardize numerical columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(india_data[numerical_cols])

# Apply PCA to reduce dimensions to 2 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
india_data['pca_1'] = pca_data[:, 0]
india_data['pca_2'] = pca_data[:, 1]
```

```python
# Step 4: Summary
summary = {
    "Total Entries": len(india_data),
    "Average Rating": india_data['rating'].mean(),
    "Average Duration": india_data['duration'].mean(),
    "Most Common Genre": india_data['genre'].mode()[0],
    "PCA Variance Explained": pca.explained_variance_ratio_.tolist()
}
```

```python
# Print summary
print(summary)
```
